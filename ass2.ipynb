```python

def random_pattern():
    return random.choice([0,1])


list={}

def random_goal(comb):
    bit=0
    goal=[]
    while bit != comb:
        rand_bit=random_pattern()
        goal.append(rand_bit)

        bit +=1
    return goal

def string_maker(comb):
    bit=0
    pattern=[]
    while bit != comb:
        rand_bit=random_pattern()
        pattern.append(rand_bit)
        bit +=1
    return pattern
```

```python
def game(comb):
    start=time.time()
    goal=random_goal(comb)
    string=string_maker(comb)
    count=0
    while string != goal:
        count +=1
        string=string_maker(comb)
    end=time.time()
    timex=end-start
    return ((f'string:{string} | time:{timex:.6f} | count:{count} | goal:{goal}'))

lenght=np.arange(2,17,1)
list={}


for comb in lenght:
    values=game(comb)

    list[comb]=values
list
    
```

```python
#V2 game

def game(comb):
    start=time.time()
    goal=random_goal(comb)
    string=string_maker(comb)
    count=0
    while string != goal:
        count +=1
        string=string_maker(comb)
    end=time.time()
    timex=end-start
    return string,timex,count,goal
```

```python
lenght=np.arange(2,17,1)

episodes=np.arange(1,31,1)
data=[]
df=pd.DataFrame(columns=['lenght','timex','count'])

for n in episodes:
    for comb in lenght:
        values=game(comb)
        data.append([comb,n,*values])


```

# Implementing mutation

```python
#V3 game+fitness

def game(comb):
    start=time.time()
    goal=random_goal(comb)
    string=string_maker(comb)
    count=0
    fit=fitness(string,comb,goal)
    while string != goal:
        count +=1
        string_n=single_mutation(string)
        fit_n=fitness_n(string_n,comb,goal)
        if fit_n>fit:
            string=string_n
            fit=fit_n
        else: string=string
        fit=fit

    end=time.time()
    timex=end-start
    return string,timex,count,goal
    
def fitness(string,comb,goal):
    fit=0
    range=np.arange(0,comb,1)
    for i in range:
        if string[i]==goal[i]:
             fit += (1/comb)
    return fit

def fitness_n(string_new,comb,goal):
    fit_n=0
    range=np.arange(0,comb,1)
    for i in range:
        if string_new[i]==goal[i]:
             fit_n += (1/comb)
    return fit_n

def single_mutation(string):
    mut_string=string.copy()
    random_digit=random.choice(range(len(string)))
    if mut_string[random_digit] == 0:
        mut_string[random_digit] +=1
    else: mut_string[random_digit] =0
    return mut_string



lenght=np.arange(2,17,1)

episodes=np.arange(1,31,1)
data=[]
df=pd.DataFrame(columns=['lenght','timex','count'])

for n in episodes:
    for comb in lenght:
        values=game(comb)
        data.append([comb,n,*values])
        
df=pd.DataFrame(data, columns=['comb','n','string','timex','count','goal'])

pivot1=df.pivot(index='comb',columns='n',values='timex')
pivot2=df.pivot(index='comb',columns='n',values='count')



```

```python
plt.subplot(1, 2, 1)
pivot1.mean(axis=1).plot(figsize=(10,5))
plt.xlabel('Combination Index')
plt.ylabel('Average time per combo')
plt.title('Time')

plt.subplot(1, 2, 2)
pivot2.mean(axis=1).plot(figsize=(10,5))
plt.xlabel('Combination')
plt.ylabel('Average attempts per combo')
plt.title('Attempts')

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

mean1 = pivot1.mean(axis=1)
mean2 = pivot2.mean(axis=1)

ax1.plot(mean1, color='b')
ax1.set_xlabel('Combination')
ax1.set_ylabel('Average time', color='b')

ax2 = ax1.twinx()
ax2.plot(mean2, color='r')
ax2.set_ylabel('Average attemepts', color='r')


plt.show()
```

The combination pattern grows pretty smoothly while the runtime shows an higher variance.
